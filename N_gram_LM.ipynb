# Part 1: N-gram Language Models

## Part 1.1: Preprocessing
Implement the function *preprocess* that iterates over all the data files and creates a single vocabulary, containing all the tokens in the data. Our token definition is a single UTF-8 encoded character. So, the vocabulary list is a simple Python list of all the characters that you see at least once in the data. The vocabulary should include the `<start>` and  `<end>` tokens.

Note - do NOT lowercase the sentences.

```python
def preprocess() -> list[str]:
  '''
  Return a list of characters, representing the shared vocabulary of all languages
  '''
  vocab = []
  vocab_set = set()
  ########################################
  all_files = os.listdir('data')
  for filename in all_files:
    if filename.endswith(('.json')):
      filepath = os.path.join('data', filename)
      #print("This is the file path", filepath)
      with open(filepath, 'r', encoding='utf-8') as f:
        data = json.load(f) # Load the JSON content
        if 'tweet_text' in data:
            for tweet_content in data['tweet_text'].values():
                for char in tweet_content:
                    vocab_set.add(char)
  vocab = sorted(list(vocab_set))
  vocab.insert(0, '<start>')
  vocab.append('<end>')
  ########################################
  return vocab
# Must deal with the case where we have an error
print(preprocess())
```

## Part 1.2: Building the Language Model
Implement the function *build_lm* that generates a language model from a textual corpus. The function should return a dictionary (representing a model) where the keys are all the relevant *n*-1 sequences, and the values are dictionaries with the *n*_th tokens and their corresponding probabilities to occur. To ensure consistent probabilities calculation, please add n-1 `<start>` tokens to the beginning of a tweet and one `<end>` token at the end. For example, for a trigram model (tokens are characters), it should look something like:

{ "ab":{"c":0.5, "b":0.25, "d":0.25}, "ca":{"a":0.2, "b":0.7, "d":0.1} }

which means for example that after the sequence "ab", there is a 0.5 chance that "c" will appear, 0.25 for "b" to appear and 0.25 for "d" to appear.

Note - You should think how to add the add-one smoothing information to the dictionary and implement it.

Please add the `<unk>` token with $p(<unk>)=1/|V|$ to the LM if building a smoothed LM.

```python
def build_lm(lang: str, n: int, smoothed: bool = False) -> dict[str, dict[str, float]]:
  '''
  Return a language model for the given lang and n_gram (n)
  :param lang: the language of the model
  :param n: the n_gram value
  :param smoothed: boolean indicating whether to apply smoothing
  :return: a dictionary where the keys are n_grams and the values are dictionaries
  '''
  LM = {}
  ########################################
  all_files = os.listdir('data')
  target_filepath = None
  for filename in all_files:
    if filename == lang + ".json":
      target_filepath = os.path.join('data', filename)
      break

  if not target_filepath:
      # If language file is not found, return empty LM
      return LM

  with open(target_filepath, 'r', encoding='utf-8') as f:
    json_data = json.load(f)

    if 'tweet_text' in json_data:
        LM['<end>'] = {}
        for tweet in json_data['tweet_text'].values():
          tokens = ['<start>'] * (n-1) + list(tweet) + ['<end>']

          for i in range(len(tokens) - (n - 1)):
            outer_key = ''.join(tokens[i:i + (n-1)])
            inner_key = tokens[i+ (n-1)]

            if outer_key not in LM:
              LM[outer_key] = {}

            if inner_key in LM[outer_key]:
              LM[outer_key][inner_key] += 1
            else:
              LM[outer_key][inner_key] = 1

  # Now, handle smoothing and convert counts to probabilities
  if smoothed:
    vocab = preprocess() # Get the full character vocabulary from preprocess()

    vocab = set(vocab)
    vocab_size = len(vocab)
    # Iterate through each context (outer_key) in the LM
    for outer_key in list(LM.keys()):
      current_inner_dict = LM[outer_key]

      # Calculate the sum of observed counts for this context
      total_observed_for_context = sum(current_inner_dict.values())

      # The denominator for smoothed probabilities includes pseudo-counts for all vocab characters
      denominator = total_observed_for_context + vocab_size

      # Create a new dictionary for smoothed probabilities for this context
      smoothed_probabilities = {}

      # Assign smoothed probabilities for each character in the vocabulary
      for char in vocab:
        count = current_inner_dict.get(char, 0) # Get observed count, or 0 if not seen
        smoothed_probabilities[char] = (count + 1) / denominator

      smoothed_probabilities["<unk>"] = 1 / denominator
      # Replace the original inner dictionary with the smoothed probabilities
      LM[outer_key] = smoothed_probabilities

  else: # If not smoothed, just normalize the raw counts to probabilities
    for outer_key, inner_dict in LM.items():
      total_count = sum(inner_dict.values())
      if total_count == 0:
          continue # Should ideally not happen if context is observed
      for inner_key, count in inner_dict.items():
        LM[outer_key][inner_key] = count / total_count

  ########################################
  return LM

# Test the function with formatted output
dictionary = build_lm('es', 3, True)
vocab = preprocess()
missing = set(vocab) - set(dictionary.keys())
print("Missing contexts:", missing)
#print(json.dumps(dictionary, indent=2))
print(len(dictionary))
```

## Part 1.3: Evaluation
We provide you with the `perplexity` function below. Your task is to implement the `eval` function which uses `perplexity` to evaluate a language model on a target language dataset.

```python
def eval(model: dict, target_lang: str, n: int) -> float:
  '''
  Return the perplexity value calculated over applying the model on the text file
  of the target_lang language.
  :param model: the language model
  :param target_lang: the target language
  :param n: The n-gram of the model
  :return: the perplexity value
  '''
  pp = 0
  ########################################
  total_log_prob = 0
  total_tokens = 0

  all_files = os.listdir('data')
  target_filepath = ""

  for filename in all_files:
    if filename == target_lang + ".json":
      target_filepath = os.path.join('data', filename)
      break

  if not target_filepath:
    print("Language file " + target_lang + " wasn't found")
    return float('inf')

  with open(target_filepath, 'r', encoding='utf-8') as f:
    json_data = json.load(f)

    if 'tweet_text' in json_data:
      for tweet in json_data['tweet_text'].values():

        #processed_tweet = ("<start> " * (n-1)) + tweet + " <end>"
        tokens = ['<start>'] * (n-1) + list(tweet) + ['<end>']

        for i in range(len(tokens) - (n - 1)):
          outer_key = ''.join(tokens[i:i + (n-1)])
          inner_key = tokens[i + (n-1)]

          if outer_key in model:
            if inner_key in model[outer_key]:
              prob = model[outer_key][inner_key]
            else:
              prob = model[outer_key]["<unk>"]
          else:
            some_context = list(model.keys())[0]
            prob = model[some_context]["<unk>"]

          total_log_prob += np.log(prob)
          total_tokens += 1

  pp = np.exp(- total_log_prob / total_tokens)
  ########################################
  return pp

lm = build_lm('en', 3, True)
print("English Preplexity: ",round(eval(lm, 'en', 3), 2))
print("French Preplexity: ",round(eval(lm, 'fr', 3), 2))
print("Filipino Preplexity: ",round(eval(lm, 'tl', 3), 2))
print("Holland Preplexity: ", round(eval(lm, 'nl', 3), 2))
```

## Part 1.4: Generation
Implement the *generate* function which takes a language code, *n*, the prompt (the starting text), the number of tokens to generate, and *r*, which is the random seed for any randomized action you plan to take in your implementation. The function should start generating tokens, one by one, using the language model of the given source language and *n*. The prompt should be used as a starting point for aligning on the probabilities to be used for generating the next token.

Note - The generation of the next token should be from the LM's distribution with NO smoothing.

```python
import random
def generate(lang: str, n: int, prompt: str, number_of_tokens: int, r: int) -> str:
  '''
  Generate text in the given language using the given parameters.
  :param lang: the language of the model
  :param n: the n_gram value
  :param prompt: the prompt to start the generation
  :param number_of_tokens: the number of tokens to generate
  :param r: the random seed to use
  '''
  text = ""
  random.seed(r)
  ########################################
  lm_dict = build_lm(lang, n)
  text = prompt
  # Last n-1 characters
  context = prompt[-(n-1):]
  # We extract the last n-1 tokens from the prompt
  # We then search the next best token for this by argmax lm_dict[next_token].items()
  for _ in range(number_of_tokens):
    if context in lm_dict:
      next_probs = lm_dict[context]
      max_prob = max(next_probs.values())

      # only choose tokens that are debatible candidates
      candidate_tokens = list(filter(lambda tok: next_probs[tok] >= max_prob - 0.1, next_probs))
      candidate_probs = [next_probs[tok] for tok in candidate_tokens]
      total = sum(candidate_probs)
      candidate_probs = [p/total for p in candidate_probs]
      # random choice among candidates
      next_token = random.choices(candidate_tokens, weights=candidate_probs, k=1)[0]
    else:
      next_token = '<unk>'
    if next_token == '<end>':
      break

    text += next_token
    context = context[1:] + next_token
  ########################################
  return text
```
